### CIFAR 10 veri kümesiyle veri artırma örnekleri ve görselleştirme

```python
import keras.preprocessing.image as ks
from keras.datasets import cifar10
from matplotlib import pyplot
from keras import backend as K 
import numpy as np
```

```python
#Veriyi indirme ve eğitim test kümelerini rastgele oluşturma

(X_train, y_train),(X_test, y_test)=cifar10.load_data()

X_train=X_train.astype('float32')
X_test=X_test.astype('float32')

X_train/=255
X_test/=255
```

## Orijinal görüntüler

```python
datagen=ks.ImageDataGenerator()
datagen.fit(X_test)


for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=4, seed=499):
  for i in range(0,4):
    pyplot.subplot(220 +1 +i)
    pyplot.imshow(X_batch[i])
  pyplot.show()
  break
```

### Döndürme

```python
datagen=ks.ImageDataGenerator(rotation_range=359)
datagen.fit(X_test)


for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=4, seed=499):
  for i in range(0,4):
    pyplot.subplot(220 +1 +i)
    pyplot.imshow(X_batch[i])
  pyplot.show()
  break
```

### Dikeyde kaydırma

```python
datagen=ks.ImageDataGenerator(height_shift_range=0.5)
datagen.fit(X_test)


for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=4, seed=499):
  for i in range(0,4):
    pyplot.subplot(220 +1 +i)
    pyplot.imshow(X_batch[i])
  pyplot.show()
  break
```

### Yatayda simetriğini alma

```python
datagen=ks.ImageDataGenerator(horizontal_flip=True)
datagen.fit(X_test)


for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=4, seed=499):
  for i in range(0,4):
    pyplot.subplot(220 +1 +i)
    pyplot.imshow(X_batch[i])
  pyplot.show()
  break
```
