```python
import keras.preprocessing.image as ks
```

ImageDataGenarator: https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image/ImageDataGenerator

```python
datagen=ks.ImageDataGenerator(rotation_range=40,
                              width_shift_range=0.2,
                              height_shift_range=0.2,
                              shear_range=0.2,
                              zoom_range=0.2,
                              horizontal_flip=True,
                              vertical_flip=True,
                              fill_mode='nearest')
```

```python
from keras.utils.image_utils import img_to_array, load_img
# goruntuleri yukleme
img=load_img('image.jpg')
x=img_to_array(img)
x=x.reshape((1,)+x.shape)
```

```python
#goruntu uretme
i=0
for batch in datagen.flow(x,batch_size=1,
                          save_to_dir='/content/drive/MyDrive/çalışmalarım/derin öğrenme kursu/veriartirma',
                          save_format='jpeg'):
  i+=1
  if i>50:
    break
```
